```python
# Importar a biblioteca pandas
import pandas as pd

# Carregar os arquivos CSV em dataframes
# (um dataframe é como uma tabela do Excel no Python)
df_produtos = pd.read_csv('olist_products_dataset.csv')
df_pedidos = pd.read_csv('olist_orders_dataset.csv')
df_clientes = pd.read_csv('olist_customers_dataset.csv')
```

```python
# Ver as 5 primeiras linhas da tabela de clientes
print("Clientes:")
print(df_clientes.head())

# Ver informações gerais, como tipos de dados e se há dados faltando
print("\nInformações dos Clientes:")
df_clientes.info()
```

```python
# Importar as bibliotecas de gráficos
import matplotlib.pyplot as plt
import seaborn as sns

# Contar quantas vezes cada categoria de produto aparece
contagem_categorias = df_produtos['product_category_name'].value_counts().head(10) # Pegando as 10 maiores

# Criar o gráfico
plt.figure(figsize=(10, 6)) # Define o tamanho do gráfico
sns.barplot(x=contagem_categorias.values, y=contagem_categorias.index, palette='viridis')
plt.title('Top 10 Categorias de Produtos Mais Vendidas')
plt.xlabel('Número de Produtos Vendidos')
plt.ylabel('Categoria do Produto')
# SALVAR a figura no arquivo, garantindo que as margens se ajustem para não cortar texto.
plt.savefig('produtos_mais_vendidos.png', bbox_inches='tight')
plt.show() # Mostra o gráfico
```

```python
# Contar clientes por estado e pegar os 10 maiores
clientes_por_estado = df_clientes['customer_state'].value_counts().head(10)

# Criar o gráfico
plt.figure(figsize=(10, 6))
sns.barplot(x=clientes_por_estado.index, y=clientes_por_estado.values, palette='mako')
plt.title('Top 10 Estados com Mais Clientes')
plt.xlabel('Estado')
plt.ylabel('Número de Clientes')
plt.savefig('estados_com_mais_clientes.png', bbox_inches='tight')
plt.show()
```

```python
# --- ANÁLISE 3: VALOR MÉDIO DO PEDIDO ---

# Passo 1: Carregar o arquivo de dados de pagamentos.
# Certifique-se de que você já fez o upload do arquivo 'olist_order_payments_dataset.csv' para o Colab.
df_pagamentos = pd.read_csv('olist_order_payments_dataset.csv')

# Passo 2: Calcular a média da coluna 'payment_value'.
# A função .mean() do Pandas calcula a média de uma coluna de números.
valor_medio = df_pagamentos['payment_value'].mean()

# Passo 3 (Opcional, mas recomendado): Formatar o número para ficar mais bonito.
# Vamos arredondar para 2 casas decimais, como fazemos com dinheiro.
valor_medio_formatado = round(valor_medio, 2)

# Passo 4: Mostrar o resultado na tela com uma frase explicativa.
print(f"Pergunta: Qual o valor médio de um pedido na loja?")
print(f"Resposta: O valor médio de um pedido é de R$ {valor_medio_formatado}")
```
